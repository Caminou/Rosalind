```python
### Mendel's First Law

def Mendel(k,m,n):
    total= k+m+n
    Paa_aa = (n/total)*((n-1)/(total-1))
    PAa_aa = (1/2)*((m/total)*(n)/(total-1))
    Paa_Aa = (1/2)*((n/total)*(m/(total-1)))
    PAa_Aa = (1/4)*((m/total)*(m-1)/(total-1))
    Paa = Paa_aa+PAa_aa+PAa_Aa+Paa_Aa
    prob = 1-Paa
    print(prob)

    
Mendel(20,26,30)
```
